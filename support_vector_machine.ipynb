```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
df = pd.DataFrame(iris.data , columns = iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
df.head()
```

```python
df['flower name'] = df.target.apply(lambda x: iris.target_names[x])
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
df0 = df[df.target==0]
df1 = df[df.target==1]
df2 = df[df.target==2]
```

```python
df0.head()
```

```python
df1.head()
```

```python
df2.head()
```

```python
plt.xlabel('sepal length(cm)')
plt.ylabel('sepal width(cm)')
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],marker='+',color='orange')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],marker='.',color='purple')
```

```python
plt.xlabel('petal length(cm)')
plt.ylabel('petal width(cm)')
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'],marker='+',color='orange')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],marker='.',color='purple')
```

```python
x = df.drop(['target','flower name'],axis='columns')
y = df.target
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.predict([[6.3,2.9,5.6,1.8]])
```
